    - 点の分布から、線を引く
    - 教師あり学習である

## データをベクトルに変換
- ex: 白黒画像
    - しろ: 0, くろ：1でピクセルごとに数字を降ればいい
    - 3 * 3のピクセルであれば9次元のベクトルになる
    - ベクトルの変換の仕方は多様であり、その変換の仕方を素性選択（特徴量/feature）という
- 文章をどうやってベクトルに変換するのか
    - **Bag of Words**という素性選択を利用する
    - 学習データに登場する全単語を覚える
    - それぞれの文章において、どの単語が登場したかをベクトル化
    - ref: ノート
