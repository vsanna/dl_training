````python
# 7章 畳み込みニューラルネットワーク
- Convolutional Neural Network CNN

## 全体の構造
- これまでの各層のつながり: 全結合(Affine)
    - 隣接する層の全てのニューロン間で結合がある
    - ex: (input) -> (affine -> ReLU) -> (affine -> ReLU) -> (affine -> Softmax)
- CNNでは新たに畳み込みそうとpooling層が追加される
    - ex: (input) -> (conv -> ReLU -> pool) -> (conv -> ReLU -> pool) -> (conv -> ReLU) -> (affine -> ReLU) -> (affine -> Softmax)
    - pooling層は省略されることもある(3層目)
    - **出力に近いそうではこれまでのaffine -> softmaxが使われる**

## 畳み込み層
- 全結合層の問題点
    - データの形状が無視されること
    - ex: 画像データは本来(1, 28, 28) = 1チャンネル, 縦28ピクセル, 横28ピクセルの3次元データ形状をしているが、それを1列に並べた784個のデータとしてaffineライヤーに送っていた
- そこで畳み込みでは形状を維持する
    - 畳み込み層の入出力データを**特徴マップ(feature map)**と呼ぶ
- 畳み込み演算
    - 入力マップにたいして、フィルター（カーネルとも言う）を掛けて、違うマップに変換する
        - フィルターがこれまでの重みにあたる
    - パディング、ストライド ... p210, p212がわかりやすい
    - パディングの目的: 出力サイズを調整するため. 
            - 入力サイズ(4, 4)に対して出力サイズも(4,4)のままにできれば、畳込みができなくなるような場面が生まれない
- 複数チャンネルの場合(= 色がある)、各チャンネルごとにその色の値を持つ
    - RGBの3チャンネルの場合を考えるとわかりやすい
    - つまり、１つのピクセルが3チャンネルにまたがって存在しており、rチャンネルにはRの、gチャンネルにはGの、bチャンネルにはBの値を格納させることになる
- 3次元データの畳込み
    - 3チャンネルの入力マップ * 3チャンネルのフィルターを計算し、チャンネルごとの検算けっかを加算して一つの出力を出す
    - つまり、H*W*C ・ FH * FW * C => 1 * OH * OWとする。1チャンネルに圧縮する。
- 出力にもチャンネルを複数持たせたい場合には？
    - フィルターにoutput_channelを加え、4次元にすればいい
    
## プーリング層
- プーリングは縦横空版を小さくする演算
- ex: maxプーリング
    - ex: 2 * 2のmaxプーリングをストライド2で行った場合
    - 他にはaverageプーリングなどがある
```
array([[ 9.19647164,  7.48357488,  0.15479468,  6.94480263],
            [ 1.50246433,  1.89127092,  3.58384519,  8.03146507],
            [ 5.71431052,  5.60533691,  1.35085372,  3.15126297],
            [ 0.23416017,  8.59571393,  7.31881557,  1.94555012]])
↓
[[9, 8],
  [8, 7]]
```
- プーリングの特徴
    1. 学習するパラメータを持たない。処理だけ
    2. チャンネル数は変化しない
    3. 微小な位置変化に対してロバスト ... 入力値が少しずれていても似たようなけっかを返す

## Convolutional / Pooling
- CNNでは各層を流れるデータは4次元のデータになる
- im2col ... image to columnというトリックで実装を簡潔化する
- 実装は省略する

## CNNの可視化
- 1層目の重みの可視化
    - フィルターは1チャンネルのグレー画像として可視化できる
    - [ ] 結局フィルターは一体どういうものなのか。
        - 特殊な配置で穴の空いたザル？そこを通れるものは◯◯な特徴を持っているみたいな？
        - ある特徴を持っている場合にだけ反応するようなイメージ？
    - [ ] ここで可視化したグレー画像は一体なに？
- 層を重ねるごとにより抽象的な特徴に反応するようになる

````
